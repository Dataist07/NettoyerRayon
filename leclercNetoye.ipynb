```python
import csv
import pandas as pd
from datetime import date

month = '2'
today = date.today()
year = today.year


dfSupermarket= pd.read_csv(f'/Users/minh/Desktop/Comparator/Scrap/Data_base/{year}/{month}/drivesScraped/list_drives_leclercScraped.csv', dtype={"department": str})

for i,supermarket in enumerate(dfSupermarket['supermarket']):
    department = dfSupermarket.iloc[i]["department"]
    city = dfSupermarket.iloc[i]["city"]
    dateScraped= dfSupermarket.iloc[i]["date_scraped"]
    nom_drive = dfSupermarket.iloc[i]["nom_drive"]

    dfProduits = pd.read_csv(f'/Users/minh/Desktop/Comparator/Scrap/Data_base/{year}/{month}/{department}/{city}/{nom_drive}/{dateScraped}/{supermarket}_produits.csv')
    # 1
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Fruits Légumes', 'Fruits, légumes') 
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon primeur'), 'rayon_principal'] = 'Fruits, légumes'
    #2
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Viandes Poissons', 'Viandes, poissons')
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon boucher'), 'rayon_principal'] = 'Viandes, poissons'
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon poissonnier'), 'rayon_principal'] = 'Viandes, poissons'
    #3
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Charcuterie Traiteur ', 'Charcuterie, traiteur')
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon Traiteur'), 'rayon_principal'] = 'Charcuterie, traiteur'
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon charcutier'), 'rayon_principal'] = 'Charcuterie, traiteur'
    #4
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Pains Pâtisseries', 'Pain, pâtisserie')
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon boulanger pâtissier'), 'rayon_principal'] = 'Pain, pâtisserie'
    #5
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Laitier Oeufs Végétal', 'Produits laitiers, oeufs, fromages')
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Mon marché frais') & (dfProduits['rayon_secondaire'] == 'Mon fromager'), 'rayon_principal'] = 'Produits laitiers, oeufs, fromages'
    #6
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Bio à prix E.Leclerc', 'Bio')
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Bébé') & (dfProduits['rayon_secondaire'] == 'Le BIO'), 'rayon_principal'] = 'Bio'
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Epicerie sucrée') & (dfProduits['rayon_secondaire'] == 'Epicerie sucrée Bio'), 'rayon_principal'] = 'Bio'
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Epicerie salée') & (dfProduits['rayon_secondaire'] == 'Epicerie salée Bio'), 'rayon_principal'] = 'Bio'
    #7
    #8
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Epicerie sucrée', 'Épicerie sucrée')
    #9
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Epicerie salée', 'Épicerie salée')
    #10
    #11
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Épicerie salée') & (dfProduits['rayon_secondaire'] == 'Saveurs du monde'), 'rayon_principal'] = 'Produits du monde'
    #12
    dfProduits.loc[(dfProduits['rayon_principal'] == 'Épicerie sucrée') & (dfProduits['rayon_secondaire'] == 'Nutrition et Bien-être'), 'rayon_principal'] = 'Nutrition, végétale'
    #13
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Univers de bébé', 'Bébé')
    #14
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Hygiène Beauté', 'Hygiène, beauté')
    #15
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Entretien Nettoyage', 'Entretien, nettoyage')
    #16
    dfProduits['rayon_principal'] = dfProduits['rayon_principal'].replace('Animalerie', 'Animal')

    # Save the modified DataFrame (optional)
    dfProduits.to_csv(f'/Users/minh/Desktop/Comparator/Scrap/Data_base/{year}/{month}/{department}/{city}/{nom_drive}/{dateScraped}/{supermarket}_produits.csv', index=False)  # Adjust path as needed

```
